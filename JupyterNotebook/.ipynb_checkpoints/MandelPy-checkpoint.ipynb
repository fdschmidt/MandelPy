# MandelPy

This Notebook contains a Python script to calculate and visualize Mandbrot sets. You can zoom in or out by changing the relevant input parameters and running the main code again.

##### Import necessary libraries

```python
import matplotlib.pyplot as plt     
import matplotlib.colors as colors  
import matplotlib.ticker as ticker     
from datetime import datetime
import numpy as np
```

### Input Parameters

```python
# Set the size of the computational domain
xMin = -2.55
xMax = 0.75
yMin = -1.6
yMax = 1.6

# Set the number of cells in each direction
xCells = 1500
yCells = 1500

# Iteration maximum
itMax = 100

# Threshold to determine whether |z_n| stays bound
threshold = 2.0

# Set colour map
colourMap = "inferno"
```

### Functions

##### CalculateIterations

```python
# This function calculates the value of z after itMax iterations
def CalculateIterations(cNumber):
    # Set the initial value of z equal to the complex number
    z = cNumber
    
    # Iterate over itMax
    for ii in range(itMax):
        
        # Stop if threshold surpassed
        if abs(z) > threshold:
            # Return number of iterations necessary to reach threshold
            return ii
        
        # Update complex number
        z = z*z + cNumber

    # Return if threshold is never reached
    return 0
```

##### CalculateMandelbrotSet

```python
# This function determines the value of 
def CalculateMandelbrotSet(xDomain, yDomain, mandelbrot):
    # Loop over rows (y values)
    for iY in range(len(yDomain)):
        
        # Loop over all columns (x values)
        for iX in range(len(xDomain)):
            
            # Generate complex number
            cNumber = np.complex(xDomain[iX], yDomain[iY])
            
            # Update mandelbort set
            mandelbrot[iY][iX] = CalculateIterations(cNumber)
            
    return mandelbrot
```

### Main Code

##### Step 1: Generate the Grid

Values along the $x$-axis will represent the real part of the complex number while values along the $y$-axis represent the imaginary part:
$c = x + y \cdot i$

```python
xDomain = np.linspace(xMin, xMax, xCells)
yDomain = np.linspace(yMin, yMax, yCells)
```

##### Step 2: Generate empty Mandelbrot Set

```python
# Generate a 2D list of dimension [xCells, yCells] filled with zeros. This list will 
# eventually store our result
mandelbrot = np.zeros([yCells, xCells])
```

##### Step 3: Calculate Mandelbrot Set

```python
# Call the CalculateMandelbrotSet
mandelbrot = CalculateMandelbrotSet(xDomain, yDomain, mandelbrot)
```

### Visualization

##### Display Plot

```python
# Initialise figure
fig = plt.figure(figsize=(13, 13), dpi=60)
pl = fig.add_subplot(111)

# Axis limits
pl.set_xlim(xMin, xMax)
pl.set_ylim(yMin, yMax)

# Labels
pl.set_title(r"MandelPy")
pl.set_xlabel(r"$x$")
pl.set_ylabel(r"$y$")

# Colour boundaries
bounds = np.linspace(0, 100, 300)
norm = colors.BoundaryNorm(boundaries=bounds, ncolors=256)

# Colormesh plot
plot = pl.pcolormesh(xDomain, yDomain, mandelbrot[:,:], norm=norm, cmap=colourMap)

# Equalize x and y axis
plt.gca().set_aspect('equal', adjustable='box')

# Show plot
plt.show()
```

##### Save Image

```python
# Determine timestamp for file name
timeStamp = datetime.now().strftime('%Y%m%d-%Hh%Mmin')

# Set file name
fileName = "MandelPy_{0}.png".format(timeStamp)

# Store plot in file
fig.savefig(fileName)

# Close plot
plt.close(fig)
```
